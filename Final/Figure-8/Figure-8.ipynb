```python
%matplotlib inline
from os.path import join
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
colors = ['#DCEBC8', '#8AC8C2', '#3D9FBA', '#4165A0']
# colors = mpl.cm.YlGnBu(np.linspace(0, 1, 4))
```

```python
plot_colors = [item for sublist in [[e]*3 for e in colors] for item in sublist]
```

##Load mapping file

```python
home = '/home/office-microbe-files'
map_fp = join(home, 'master_map_150908.txt')
```

```python
sample_md = pd.read_csv(map_fp, sep='\t', index_col=0, dtype=str)
sample_md = sample_md[sample_md['16SITS'] == '16S']
sample_md = sample_md[sample_md['OfficeSample'] == 'yes']
sample_md = sample_md[sample_md['Row'] == '3']
```

```python
replicate_ids = '''F2F.2.Ce.021
F2F.2.Ce.022
F2F.3.Ce.021
F2F.3.Ce.022
F2W.2.Ca.021
F2W.2.Ca.022
F2W.2.Ce.021
F2W.2.Ce.022
F3W.2.Ce.021
F3W.2.Ce.022
F1F.3.Ca.021
F1F.3.Ca.022
F1C.3.Ca.021
F1C.3.Ca.022
F1W.2.Ce.021
F1W.2.Ce.022
F1W.3.Dr.021
F1W.3.Dr.022
F1C.3.Dr.021
F1C.3.Dr.022
F2W.3.Dr.059
F3F.2.Ce.078'''.split('\n')
```

```python
reps = sample_md[sample_md['Description'].isin(replicate_ids)]
reps = reps.drop(reps.drop_duplicates('Description').index).index
sample_md.drop(reps, inplace=True)
```

##Load alpha diversity

```python
alpha_div_fp = '/home/office-microbe-files/core_div_out/arare_max1000/alpha_div_collated/PD_whole_tree.txt'
alpha_div = pd.read_csv(alpha_div_fp, sep='\t', index_col=0)
alpha_div = alpha_div.T.drop(['sequences per sample', 'iteration'])
alpha_cols = [e for e in alpha_div.columns if '1000' in e]
alpha_div = alpha_div[alpha_cols]
```

```python
sample_md = pd.concat([sample_md, alpha_div], axis=1, join='inner')
sample_md['MeanAlpha'] = sample_md[alpha_cols].mean(axis=1)
sample_md = sample_md[sample_md['MeanAlpha'].notnull()]
```

# Group city date by 2 week increments

```python
flag_md = sample_md[sample_md['City'] == 'flagstaff'][['MeanAlpha', 'Date']].copy()
flag_md.index = pd.DatetimeIndex(flag_md.Date)
flag_md = flag_md.sort_index()
```

```python
flag_dates = ['2013-06-28', '2013-07-12', '2013-07-26', 
              '2013-09-23', '2013-10-07', '2013-10-21',
              '2014-01-06', '2014-01-20', '2014-02-03',
              '2014-04-07', '2014-04-21', '2014-05-05']
```

```python
for i, j in enumerate(flag_dates):
    flag_md.loc[flag_md.index >= pd.to_datetime(j), 'Week'] = j
    flag_md.loc[flag_md.index >= pd.to_datetime(j), 'Period'] = i//3
```

```python
with plt.rc_context(dict(sns.axes_style("darkgrid"),
                         **sns.plotting_context("notebook", font_scale=2))):
    plt.figure(figsize=(20, 10))
    
    ax = sns.violinplot(x="Week", y="MeanAlpha", data=flag_md, palette=plot_colors)
    xlabels = ['{0} (n={1})'.format(date, n) for date, n in flag_md.groupby('Week').agg('count')[['Date']].itertuples()]
    ax.set_xticklabels(xlabels, rotation=45)
    ax.set_ylim(-20, 120)
    handles = [mpl.patches.Patch(color=e) for e in colors]
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), labels=['Summer', 'Fall', 'Winter', 'Spring'], handles=handles)
    ax.set_title('Flagstaff')
    ax.set_ylabel('Phylogenetic Diversity')
    ax.set_xlabel('')
    plt.xticks(np.linspace(-.5, 10.6, 12))
    plt.tight_layout()
    plt.savefig('figure-8-flagstaff.svg', dpi=300)
```
